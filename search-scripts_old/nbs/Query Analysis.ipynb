```python
import os
import glob
import operator
import pandas as pd
```

```python
def convert_to_csv(filename, csv_filename):
    df = pd.read_excel(filename)
    df.to_csv(csv_filename, encoding='utf-8', index=False)
```

```python
def parse_keywords(keywords):
    keywords = keywords.split(',')
    if len(keywords) == 1:
        query = keywords[0].lower()
        if 'and' in query or 'or' in query:
            return []
    
    queries = []
    for keyword in keywords:
        query = keyword.lower().replace('"', '')
        query = query.strip(' ()').strip()
        queries.append(query)
    
    return queries
```

```python
def parse_query_file(filename):
    csv_filename = filename.replace('.xlsx', '.csv')
    if not os.path.exists(csv_filename):
        print(filename)
        convert_to_csv(filename, csv_filename)
    
    df = pd.read_csv(csv_filename)
    df = df[df['ANY'].notnull()]['ANY']
    queries = df.apply(lambda keywords: parse_keywords(keywords))

    return queries
```

```python
queries = {}
filenames = glob.glob("data/queries/*.xlsx")
for filename in filenames:
    print(filename)
    data = parse_query_file(filename)
    for data_tuple in data.iteritems():
        for query in data_tuple[1]:
            if query not in queries:
                queries[query] = 0

            queries[query] += 1
```

```python
sorted_queries = sorted(queries.items(), key=operator.itemgetter(1), reverse=True)
sorted_queries = sorted_queries[1:]
```

```python
total_unique_queries = len(sorted_queries)

total_query_count = 0
for query in sorted_queries:
    total_query_count += query[1]
    
top_queries_count = 0
for query in sorted_queries[:10000]:
    top_queries_count += query[1]
```

```python
print("Total unique queries: %d" % total_unique_queries)
print("Total query count: %d" % total_query_count)
print("Top queries count: %d" % top_queries_count)
print("Top queries percentage: %d" % round(100*top_queries_count/total_query_count))
```
